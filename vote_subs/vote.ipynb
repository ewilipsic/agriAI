```python
import pandas as pd
dfs = []
for path in ["guy.csv","submission.csv","submission (1).csv"]:
    dfs.append(pd.read_csv(path))

outputs = []

for i in range(dfs[0].shape[0]):
    counts = {"RPH":0,"Blast":0,"Aphid":0,"Rust":0}
    id = dfs[0].iloc[i]["Id"]
    for df in dfs:
        counts[df[df["Id"] == id].iloc[0]["Category"]] += 1

    final_label = "RPH"
    for label,count in counts.items():
        if(count > counts[final_label]): final_label = label

    outputs.append((id,final_label)) 
```

```python
with open("voted_submission/submission.csv","w") as f:
    f.write("Id,Category\n")
    for id,label in outputs:
        f.write(f"{id},{label}\n")
```
