```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

guy = pd.read_csv('guy.csv')
sub = pd.read_csv('submission.csv')

guy = guy.sort_values(by="Id")
sub = sub.sort_values(by="Id")

# Create confusion matrix
cm = confusion_matrix(guy['Category'], sub['Category'])

# Plot it
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted (submission)')
plt.ylabel('Actual (guy)')
plt.title('Confusion Matrix')
plt.show()

# Optional: Get the labels for the axes
labels = sorted(guy['Category'].unique())
print(f"Classes: {labels}")
print(f"\nConfusion Matrix:\n{cm}")
```
